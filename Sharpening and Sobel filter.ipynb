```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import lena
from scipy.ndimage.filters import convolve
import scipy.ndimage as nd
```

# Image Filter : Sharpening / Gauss_highpass

```python
src_image = lena()
```

```python
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])
highpass = convolve(src_image, kernel)
plt.title('simple high-pass')
plt.imshow(highpass, cmap=plt.cm.gray)
```

# Image Filter :  Sobel Operator

```python
ima = np.zeros((256,256))
ima[64:-64, 64:-64] = 1
ima = nd.rotate(ima, 15, mode='constant')
ima = nd.gaussian_filter(ima, 8)
plt.imshow(ima)
```

```python
sx = nd.sobel(ima, axis=0)
sy = nd.sobel(ima, axis=1)
sob = np.sqrt(sx*sx + sy*sy)

plt.subplot(131)
plt.imshow(sx)

plt.subplot(132)
plt.imshow(sy)

plt.subplot(133)
plt.imshow(sob)

plt.gcf().set_figwidth(15)
```
