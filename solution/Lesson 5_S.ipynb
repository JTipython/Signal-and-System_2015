```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import lena
from scipy.ndimage.filters import convolve
import scipy.ndimage as nd
```

```python
src_image = lena()
```

# Q1: 如何利用gaussian filter來達成gaussian high-pass filter的影像?

```python
# gauss_highpass = original_data - gauss_lowpass
igf = nd.gaussian_filter(src_image, 3)
igf = src_image - igf
plt.title('gaussian high-pass')
plt.imshow(igf, cmap=plt.cm.gray)
```

# Q2: 參考sobel教學，改用lena影像得到下面三張結果圖。

```python
sx = nd.sobel(src_image, axis=0)
sy = nd.sobel(src_image, axis=1)
sob = np.sqrt(sx*sx + sy*sy)

plt.figure(figsize=[15,5])
plt.subplot(131)
plt.title('sobel x direction')
plt.imshow(sx, cmap=plt.cm.gray)

plt.subplot(132)
plt.title('sobel y direction')
plt.imshow(sy, cmap=plt.cm.gray)

plt.subplot(133)
plt.title('sobel filter')
plt.imshow(sob, cmap=plt.cm.gray)
plt.colorbar()
```

# Q3: 試著用之前學過的median filter來增強邊緣。

```python
sobmfed1 = nd.median_filter(sob, 5)
sobmfed2 = nd.median_filter((sob> 255 * 0.5), 5)
# 255可換，只要呈現的結果有增強邊緣即可
plt.figure(figsize=[15,5])
plt.subplot(121)
plt.imshow(sobmfed1, cmap=plt.cm.gray)

plt.subplot(122)
plt.imshow(sobmfed2, cmap=plt.cm.gray)
```
