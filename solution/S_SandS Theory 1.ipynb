```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Q1: 將兩個旋波相加後，並畫在一起。

```python
t = np.arange(0, 1, 0.001)

y1 = np.sin(2*np.pi*5*t)
y2 = 0.7 * np.sin(2*np.pi*5*t + 1*np.pi)

plt.plot(y1) # blue
plt.plot(y2) # green
plt.plot(y1 + y2) # red
```

# Q2: 嘗試取樣的不同，會出現怎樣的變化。

```python
t1 = np.arange(0, 1, 0.02)

y1 = np.sin(2*np.pi*5*t1)

plt.plot(y1)
```

# Q3: 試著用不同的量化數模擬出下面的結果。

```python
t = np.arange(0, 1, 0.001)
y = np.sin(2*np.pi*t)
y2 = (y*1.5).astype(int)
y4 = (y*3.5).astype(int)

plt.plot(y)
plt.plot(y2)
plt.plot(y4 / 3.0)
plt.legend(['sine', '1-bit', '3-bit'])
```
